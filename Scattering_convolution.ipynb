# Physics of Ultrasound Imaging: Scattering and Convolution

This document explores the two fundamental physical processes that govern how ultrasound images are formed: **Scattering** (tissue interaction) and **Convolution** (system processing).


## Scattering (Dispersion)

Scattering is the physical phenomenon responsible for the internal "texture" (parenchyma) of organs seen in an ultrasound.

### Key Concepts
* **Definition**: Scattering occurs when sound waves encounter structures with dimensions equal to or smaller than the wavelength ($\lambda$), or irregular interfaces. Energy is dispersed in multiple directions rather than reflected as a single echo.
* **Diffuse Scattering**: Occurs on rough surfaces. It is crucial because it ensures signal returns to the transducer regardless of the beam's angle.
* **Rayleigh Scattering**: A specific type occurring with microscopic structures like red blood cells. The intensity increases with the fourth power of frequency ($f^4$).



> **The Result: Speckle**
> The grainy appearance of ultrasound images is called "speckle." It is not electronic noise but the result of constructive and destructive interference of thousands of small scattered signals.



## The Point Spread Function (PSF)

Before understanding convolution, we must define the **PSF**:
* It represents the system's response to a single, ideal point scatterer.
* Since a transducer cannot produce an infinitesimal point, it creates a "blur" or "spot" characterized by **length** (pulse duration) and **width** (beam focusing).



## Convolution (Image Formation)

**Convolution** is the mathematical model describing how the physical ultrasound pulse interacts with human tissue to create the final image.

### The Mathematical Model
The final image $I(z)$ is the convolution of the tissue's scattering map $T(z)$ and the system's $PSF(z)$:

$$I(z) = T(z) * PSF(z)$$



### Impact on Image Quality
| Feature | Dependency |
| :--- | :--- |
| **Blurring** | Larger PSFs cause more "smearing" of tissue details. |
| **Axial Resolution** | Improved by higher frequencies (shorter pulses/PSF). |
| **Lateral Resolution** | Improved by narrower beam focusing (thinner PSF). |


# Time Domain Analysis of Continuous Time Systems

## Impulse Response

The formula of impulse response: 
$$h_\tau = H(\delta_\tau)$$

* $\delta_\tau$ (input): Represents the impulse sends at t time;
* H (the system): It's the operator who represents the system (ecograph);
* $h_\tau$ (output): It's the impulse response, what comes out of the system when you put a dry pulse in.

Example

- $\delta_\tau$: very short electrical pulse given by a piezoelectric crystal;
- H: includes everything that sound goes through (fabric, crystal);
- $h_\tau$: wave packet emitted by the crystal. The narrower this package is (small wavelength) the sharper the image will be.

- t: first pulse generated
- $\tau$: second delayed pulse generated 

#### The role of t in the formula

$$h_\tau(t) = H(\delta_\tau(t))$$

* t at left -> is for a fix number. What I get as output after a few seconds
* t at right -> represents the entire duration of the oscillation. From its emission via the pulse until its complete dispersion.

Example with $\tau$=0 
$$h_0(1) = H(\delta_0(3))$$

## Time invariance

The basic concept is simple: the physical properties of the system do not change over time.

* If you send a pulse at time t=0, you get a certain response h(t , 0);
* If you send the same pulse later, at time $\tau$, you get a response h(t , $\tau$) who has the identical set shape, only translated in time.

## Linearity and extended linearity

#### Linearity

A system S can be considered linear if it satisfies:

1. **Homogeneity**
   - If y = Sx, then ay = S(ax)
   - **What does it mean**: if you double the intensity of the pulse you send into the body (the input ax), the echo you receive (the output ay) will be exactly twice as loud;
   - **Why it's important**: If the system was not homogeneous by increasing the power of the ultrasound machine, the image could distort or change color unpredictably instead of simply becoming brighter.

2. **Superposition**
   - if $y_1 + y_2 = S(x_1 + x_2)$
   - **What does it mean**: If you send two pulses $(x_1 and x_2)$ and they hit two different points, the final image will simply be the sum of the two single responses;
   - **Why it's important**: Responses do not hinder each other. If one point in the liver produces an echo and a nearby point produces another, the ultrasound machine sees them both added together without creating strange interference that alters the shape of the wave packet.
   

#### Extended Linearity

1. **Summation**
   - $\sum_{n} a_n y_n = S \left( \sum_{n} a_n x_n \right)$
   - **What doest it mean**: If you have many small tissue reflection centers (the scatterers), each with its intensity $(a_n)$ and its response $(y_n)$, the ultrasound will see the sum of all these contributions at once;
   - **Why it's important**: The system operator S and the summation $\sum$ can be swapped places. This means that analyzing the response to the sum of tissues is the same as adding the responses of individual tissues.

2. **Integration**
   - $\int_{-\infty}^{\infty} a(\tau)y(t - \tau) d\tau = S \left( \int_{-\infty}^{\infty} a(\tau)x(t - \tau) d\tau \right)$
